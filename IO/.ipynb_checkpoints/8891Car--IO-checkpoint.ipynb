```python
import sqlite3

try:
    conn = sqlite3.connect('C:\\Users\\BIG DATA\\Desktop\\car_spec.db')
    cursor = conn.cursor()
#     o = cursor.execute('Select brandName,type From spec limit 1')
#     print o.fetchone()
#     print(o[0])
    
    
    for row in cursor.execute('Select brandName,type,my_id,years From spec;'):
        with open (u"./newBrandType2.csv",'a',encoding="utf-8") as fw:
            
            brand = row[0]
            type  = row[1]
            my_id  = row[2]
            years = row[3]

            fw.write("{},{},{},{}".format(brand,type,my_id,years)+'\n')
            print(brand,type,my_id,years)
    conn.commit()   
except UnicodeDecodeError as e:
    print(e)
    pass

finally:
    conn.close()
```
